# Libraries

# C. Decision boundaries (50 points)

<div dir="rtl" style="font-size:16px;">
<font face="B Yekan">

حال دیتاست خود را به ۳ بخش مساوی به صورت تصادفی تقسیم کنید. برای هر بخش ۳
[مدل طبقه بند نزدیک‌ترین همسایه (‌k-nearest neighbor)](https://fa.wikipedia.org/wiki/%D8%A7%D9%84%DA%AF%D9%88%D8%B1%DB%8C%D8%AA%D9%85_%DA%A9%DB%8C-%D9%86%D8%B2%D8%AF%DB%8C%DA%A9%E2%80%8C%D8%AA%D8%B1%DB%8C%D9%86_%D9%87%D9%85%D8%B3%D8%A7%DB%8C%D9%87)
 با پارامتر k={1, 25, 50} آموزش دهید (در مجموع ۹ ترکیب خواهید داشت)(برای به کارگیری مدل می توانید از کتابخانه استفاده کنید.). برای هر ترکیب،‌ مرز تصمیم‌گیری را به همراه تمام دیتاست رسم کنید و نتیجه را در یک نمودار سه در سه که ردیف‌های آن، دیتاست‌های 
 متفاوت و ستون‌های آن k های متفاوت اند،‌ نشان دهید.‌ (‌برای اینکار باید صفحه را مش‌بندی کنید و کلاس هر نقطه در مش را با طبقه بند خود پیش‌بینی کنید).
</div>

```python
X, y = make_moons(n_samples=1500, noise=0.35, random_state=42)
rng = np.random.RandomState(42)
perm = rng.permutation(len(X))
X_splits = np.array_split(X[perm], 3)
y_splits = np.array_split(y[perm], 3)
xx, yy = np.meshgrid(np.arange(-2, 3, 0.01), np.arange(-2, 3, 0.01))
fig, axs = plt.subplots(3, 3, figsize=(10, 10))
for i, k in enumerate([1, 25, 50]):
    for j in range(3):
        X_train, X_test, y_train, y_test = train_test_split(X_splits[j], y_splits[j], test_size=0.2, random_state=42)
        knn = KNeighborsClassifier(n_neighbors=k)
        knn.fit(X_train, y_train)
        Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
        Z = Z.reshape(xx.shape)
        axs[i, j].contourf(xx, yy, Z, cmap=cmap_light, alpha=0.3)
        axs[i, j].scatter(X_splits[j][:, 0], X_splits[j][:, 1], c=y_splits[j], cmap=cmap_bold, alpha=0.8)
        axs[i, j].set_title(f"Subset {j+1}, k={k}")
plt.show()
```

# D. Bias Variance Trade-off (20 points)

<div dir="rtl" style="font-size:16px;">
<font face="B Yekan">
 تفاوت مرز تصمیم‌گیری در ردیف‌ها و ستون‌ها را بیان کنید و مصالحه بین بایاس-واریانس را بر اساس نمودار توضیح دهید.
</div>

<h4>"As the arrangement of mixed english-persian letters was confusing in jupyter notebook, I wrote it in English"</h4> <br>

Firstly, we can see that the decision boundary tends to be more complex (with more curves and irregularities) when `k=1` compared to when `k` is larger. This is because when `k=1`, the KNN classifier assigns the label of the closest training point to each test point, which can lead to overfitting and a complex decision boundary that tries to fit the noise in the data. On the other hand, when `k` is larger, the decision boundary tends to be smoother and simpler because the KNN classifier takes into account a larger number of neighbors when making predictions, which helps to reduce the effects of noise and make the model more robust.

Secondly, we can see some differences in the decision boundary between the different subsets. This is likely due to the fact that the subsets were randomly split and may contain different distributions of data points. Some subsets may be easier to classify than others, and the decision boundary may need to be more complex in order to accurately capture the patterns in the data.

These observations relate to the bias-variance trade-off in machine learning. A model with high bias tends to underfit the data and have a simple decision boundary that does not capture the complexity of the underlying patterns in the data. On the other hand, a model with high variance tends to overfit the data and have a complex decision boundary that tries to fit the noise in the data. In KNN, the `k` parameter controls the bias-variance trade-off by determining how many neighbors to consider when making predictions. A smaller `k` value leads to a lower bias and higher variance, while a larger `k` value leads to a higher bias and lower variance. 

Overall, the decision boundary diagrams provide a visual representation of how the choice of `k` and the data distribution can affect the performance of the KNN classifier. By examining these diagrams, we can gain insights into the bias-variance trade-off and make informed decisions about the choice of hyperparameters in our models.
